```python
import pandas as pd

df = pd.read_csv("iris.csv")

df.head()
```

```python
import numpy as np
np.unique(df['variety'])
```

```python
df['class'] = df['variety'].replace({"Setosa": 0, "Versicolor": 1, "Virginica": 2})
```

```python
df = df.dropna()
```

```python
df = df.fillna(0)
```

```python
df = df.drop_duplicates()
```

```python
df.describe()
```

```python
df.info() 
#variety este object deoarece in pandas(fiind bazat pe C) nu exista data type string, 
# este un array de char-uri
```

```python
# lista_valori = df.stack().tolist()
# print(lista_valori)

df['x'] = df[df.columns[:4]].values.tolist()
df['x']
```

```python
df['y'] = df['class'].apply(lambda x: x+1)
df ['y']
```

```python
df.describe()
```

```python
df.to_csv("iris_ok.csv", index=False)
```
